```python
!pip install pypdf
from langchain.document_loaders import PyPDFLoader

def load_pdf(file_path):
  loader = PyPDFLoader(file_path) # Use PyPDFLoader for single file

  documents = loader.load()
  return documents

extrcted_data = load_pdf('/content/ai.pdf')
```

```python
# create text chunk

def text_split(extrcted_data):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500,chunk_overlap = 20)
  text_chunks= text_splitter.split_documents(extrcted_data)

  return text_chunks
```

```python
text_chunks = text_split(extrcted_data)
print(len(text_chunks))
```

```python
import pinecone
from langchain.vectorstores import Pinecone
import os

# Set your API key and environment using environment variables or directly in the code
os.environ["PINECONE_API_KEY"] = "64bf00ca-a93e-495e-94a9-3dd8f453c960"
os.environ["PINECONE_ENVIRONMENT"] = "us-east-1"

# Initialize Pinecone client (Corrected)
pinecone_client = pinecone.Pinecone(api_key=os.getenv("PINECONE_API_KEY"), environment=os.getenv("PINECONE_ENVIRONMENT"))

# Set the index name
index_name = "aibot"

# Ensure the index exists; if not, create it
if index_name not in pinecone_client.list_indexes().names():   # Use pinecone_client to list indexes
    pinecone_client.create_index(name=index_name, dimension=384)  # Use pinecone_client to create index

# Connect to the index
index = pinecone_client.Index(index_name)  # Use pinecone_client to get index

def clean_text(text):
    """Cleans text by removing or replacing invalid Unicode characters."""
    return text.encode('utf-8', 'surrogatepass').decode('utf-8', 'replace')

# ... (Rest of your code remains the same)

# Create document search object
# First, clean the text
cleaned_texts = [clean_text(t.page_content) for t in text_chunks]

# Check if the document already exists
for cleaned_text in cleaned_texts:
    # Generate the embedding for the text
    embedding = embeddings.embed_query(cleaned_text) # use embed_query to embed a single string

    # Query Pinecone to see if a similar vector already exists
    response = index.query(vector=embedding, top_k=1, include_metadata=True)

    # Only upsert if no similar vector exists
    if not response['matches']:
        docsearch = Pinecone.from_texts([cleaned_text], embeddings, index_name=index_name)
    else:
        print(f"Document already exists for text: {cleaned_text[:30]}...")
```

```python
docsearch = Pinecone.from_existing_index(index_name,embeddings)

query = "What are AI"

docs = docsearch.similarity_search(query,k=3)
print(docs)
```

```python
!pip install langchain

from langchain.llms import CTransformers
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

# ... (rest of your code)

# Wrap your ctransformers model for LangChain compatibility
llm = CTransformers(model=model_path,
                    model_type="llama",
                    max_new_tokens=512,
                    temperature=0.8)

q = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=docsearch.as_retriever(search_kwargs={"k":2}),
    chain_type_kwargs=chain_type_kwargs,
)

```

```python
from transformers import AutoTokenizer

# Initialize the tokenizer with a compatible LLaMA tokenizer from Hugging Face
tokenizer = AutoTokenizer.from_pretrained("hf-internal-testing/llama-tokenizer")

def truncate_input(input_text, max_length):
    tokens = tokenizer.encode(input_text, max_length=max_length, truncation=True)
    return tokenizer.decode(tokens, skip_special_tokens=True)

while True:
    user_input = input("Input Prompt (type 'exit' to quit): ")
    if user_input.lower() in ['exit', 'quit']:
        print("Exiting...")
        break

    try:
        truncated_input = truncate_input(user_input, max_length=512)
        result = q.invoke({"query": truncated_input})
        print(f"Answer: {result['result']}")
    except Exception as e:
        print(f"An error occurred: {e}")
```
